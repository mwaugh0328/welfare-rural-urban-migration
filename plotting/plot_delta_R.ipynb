```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
```

```python
mat = scipy.io.loadmat("targets.mat")
df = pd.DataFrame(mat["targets"])
```

```python
df.columns = ["wage", "rural", "var", "assets", "migration", "m_migration", "m_migration2",
             "LATE", "OLS", "var2","var3"]
```

```python
R = np.linspace(0.90,1.05,16)
```

```python
df["R"] = R
```

```python
fig, ax = plt.subplots(figsize = (16,9))

mike_blue = tuple(np.array([20, 64, 134]) / 255)

ax.plot(df.R, 100*df.migration, alpha = 0.95, color = "red", linewidth = 4, linestyle= "--"
        , label = "Experimental Data")

ax.set_ylim(25,42)
ax.set_xlim(0.90,1.05)

ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

ax.set_ylabel('Seasonal Migration Rate', fontsize = 16)
ax.set_xlabel('Real Intrest Rate', fontsize = 16)

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (16,9))

mike_blue = tuple(np.array([20, 64, 134]) / 255)

ax.plot(df.R, df.LATE - df.OLS, alpha = 0.95, color = "red", linewidth = 4, linestyle= "--"
        , label = "Experimental Data")


ax.set_xlim(0.90,1.05)

ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

ax.set_ylabel('LATE - OLS', fontsize = 16)
ax.set_xlabel('Real Intrest Rate', fontsize = 16)

plt.show()
```
