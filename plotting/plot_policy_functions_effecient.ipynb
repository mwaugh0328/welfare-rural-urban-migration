```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
import scipy.io
import seaborn as sns
```

```python
# load the effecient policy funcations

move_mat = scipy.io.loadmat( "movepolicy_effecient.mat")

low_move = pd.DataFrame(move_mat["lowz"])

med_move = pd.DataFrame(move_mat["medz"])

exp_move = pd.DataFrame(move_mat["lowz_exp"])

ce_move_mat = scipy.io.loadmat( "movepolicy.mat")

ce_low_move = pd.DataFrame(ce_move_mat["lowz"])

ce_med_move = pd.DataFrame(ce_move_mat["medz"])

ce_exp_move = pd.DataFrame(ce_move_mat["lowz_exp"])
```

### Migration Policy Functions

Plots them as a heat map. To create the ``.mat`` files run ``analyze_outcomes(x1, [], [], [], [], [], 1)`` with a linear asset grid, 200 points and 30 transatory shocks with the ``x1`` value comming from the baseline calibration

```python
alpha = 0.85

fig_path = "..\\figures\\effecient_"

asset_range = range(125,200,1)
shock_range = range(0,30,1)
```

```python
fig, ax = plt.subplots(1,2,figsize = (16,10))

cbar_ax = fig.add_axes([.93, .15, .03, 0.70])

sns.heatmap((1-ce_low_move).iloc[asset_range,shock_range], cmap="Blues",cbar = False,vmin=0,vmax=1.0, ax=ax[0], alpha = 0.90)

sns.heatmap((1-low_move).iloc[asset_range,shock_range], cmap="Blues",cbar = True,vmin=0,vmax=1.0, ax=ax[1],
            cbar_kws={'label': '\n Probability of Moving'}, 
             cbar_ax = cbar_ax, alpha = alpha)

label_list = ["Status Quo", "Effecient Allocation"]

count = 0

for xxx in ax:
    
    xxx.get_xaxis().set_visible(False)
    xxx.get_yaxis().set_visible(False)
    
    xxx.set_title(label_list[count])
    
    count = count + 1

ax[0].text(-2, 75, "Larger Asset Holdings ⟶ ",fontsize = 14, rotation = 90)

ax[0].text(0, 80, "Larger Transitory Shock ⟶ ",fontsize = 14)

fig.suptitle("Migration Policy Function: Low z, No Experience \n", fontsize = 16)

plt.savefig(fig_path + "migration_policy_low_z_both.pdf", bbox_inches = "tight", dip = 3600)
plt.savefig(fig_path + "migration_policy_low_z_both.png", bbox_inches = "tight")

plt.show()
```
